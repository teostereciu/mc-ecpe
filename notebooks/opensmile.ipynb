```python
import pandas as pd

df = pd.read_csv('/Users/teodorastereciu/Documents/bachelors-project/mc-ecpe/data/processed/processed_func_data.csv', dtype={'id':str})
df.head()
```

```python
import numpy as np

X_train_audio = np.array(df[df['split'] == 'train'].drop(columns=['filename', 'id', 'conversation_line', 'emotion', 'speaker', 'split']).values.tolist())
X_val_audio = np.array(df[df['split'] == 'val'].drop(columns=['filename', 'id', 'conversation_line', 'emotion', 'speaker', 'split']).values.tolist())
X_test_audio = np.array(df[df['split'] == 'test'].drop(columns=['filename', 'id', 'conversation_line', 'emotion', 'speaker', 'split']).values.tolist())
```

```python
from imblearn.over_sampling import RandomOverSampler

```

```python
import tensorflow as tf

def encode_labels(text_labels):
    """ One-hot encodes emotion labels. """

    labels = ['neutral', 'joy', 'surprise', 'anger', 'fear', 'disgust', 'sadness']
    num_classes = len(labels)
    label_to_index = {label: index for index, label in enumerate(labels)}
    index_to_label = {index: label for index, label in enumerate(labels)}

    idx_labels = [label_to_index[label] for label in text_labels]
    onehot_labels = tf.one_hot(idx_labels, len(labels)).numpy()
    return onehot_labels

train_labels = df[df['split'] == 'train']['emotion']
val_labels = df[df['split'] == 'val']['emotion']
test_labels = df[df['split'] == 'test']['emotion']

y_train = encode_labels(train_labels)
y_val = encode_labels(val_labels)
y_test = encode_labels(test_labels)
```

```python
import tensorflow as tf


input_layer = tf.keras.layers.Input(shape=(988,))

x = tf.keras.layers.Dense(256, activation='relu')(input_layer)
x = tf.keras.layers.Dense(128, activation='relu')(x)

outputs = tf.keras.layers.Dense(7, activation='softmax')(x)

model = tf.keras.Model(input_layer, outputs)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
print(model.summary())
```

```python
X_train_audio = np.array(df.head(100).drop(columns=['filename', 'id', 'conversation_line', 'emotion', 'speaker']).to_list())
```

```python

history = model.fit(X_train_audio, y_train, validation_data=(X_val_audio, y_val), epochs=3, batch_size=64)
```
